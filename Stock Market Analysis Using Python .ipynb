```python
#Removing noises using moving average 

gm['Open'].plot(label='No Moving Average', figsize=(15,7))
gm['MA50'] = gm['Open'].rolling(50).mean() # MA50 implies 50 samples
gm['MA50'].plot(label='MA50')
gm['MA200'] = gm['Open'].rolling(200).mean() # MA200 implies 200 samples
gm['MA200'].plot(label='MA200')
plt.legend();
```

If the moving average window size is too high we might miss certain trends. Similarly if the window size is too low we have noises in our visualization. Thus we need to have a balanced window size.

# Daily Percentage Change

1. Rt = (Pt/P(t-1))-1 Where Rt = Return at time t, Pt = Price at time t, P(t-1)=Price at time (t-1). 
2. Rt shows the percentage gain or loss
3. Rt helps in determining the volatility of the stock. The more dispersed the value of Rt the more volatile is the stock

```python
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) - 1
tesla['returns'].head()
```

```python
ford['returns'] = (ford['Close']/ford['Close'].shift(1)) - 1
ford['returns'].head()
```

```python
gm['returns'] = (gm['Close']/gm['Close'].shift(1)) - 1
gm['returns'].head()
```

```python
#Plotting histogram 

tesla['returns'].hist(bins=100, label = 'Tesla', alpha = 0.5, figsize=(15, 7))
ford['returns'].hist(bins=100, label = 'Ford', alpha = 0.5)
gm['returns'].hist(bins=100,  label = 'GM', alpha = 0.5)
plt.legend();
```

```python
#Plotting KDE

tesla['returns'].plot(kind='kde', label = 'Tesla', figsize=(15, 7))
ford['returns'].plot(kind='kde', label = 'Ford')
gm['returns'].plot(kind='kde',  label = 'GM')
plt.legend();
```

1. The higher the standard deviation of the histogram the higher volatile is the stock.
2. As we can see Tesla has the highest volatility.
3. The stocks of Ford and GM are more stable as compared to Tesla

```python
#BoxPlots
box_df = pd.concat([tesla['returns'], gm['returns'], ford['returns']], axis = 1)
box_df.columns=['Tesla Returns', 'GM Returns', 'Ford Returns']
box_df.plot(kind='box', figsize=(16,6));
```

# Cumulative Return

```python
tesla['Cumulative Return']=(1+tesla['returns']).cumprod()
tesla.head()
```

```python
ford['Cumulative Return']=(1+ford['returns']).cumprod()
gm['Cumulative Return']=(1+gm['returns']).cumprod()
```

```python
tesla['Cumulative Return'].plot(label='Tesla', figsize=(15, 7))
ford['Cumulative Return'].plot(label='Ford')
gm['Cumulative Return'].plot(label='GM')
plt.title('Cumulative Return vs Time')
plt.legend();
```

The TESLA stock showed the highest return for $1 invested.
